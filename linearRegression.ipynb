**Building a Linear Regression Model to Predict Fuel Efficiency (MPG) Using the Auto MPG  Dataset**

```python
import zipfile
import os

# Unzip the file
with zipfile.ZipFile("/content/auto+mpg.zip", 'r') as zip:
    zip.extractall("/content/auto_mpg_zip")

import pandas as pd

column_names = [
    "mpg", "cylinders", "displacement", "horsepower", "weight",
    "acceleration", "model_year", "origin", "car_name"
]
# get data from desired file and create a dataframe
df = pd.read_csv("//content/auto_mpg_zip/auto-mpg.data",
                 delim_whitespace=True,
                 names=column_names,
                 na_values="?")

```

```python
df.head()
```

**Data Preprocessing**

```python
df.isnull().sum()
```

```python
#To fill missing horsepower values with the median horsepower of cars having the same number of cylinders
df['horsepower'] = df.groupby('cylinders')['horsepower'].transform(lambda x : x.fillna(x.median()))
df['horsepower'].isnull().sum()
# independent and dependent variables are identified
x = df.drop(["car_name","mpg"],axis = 1)
y = df["mpg"]
```

```python
# import of different modules
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, r2_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 41)
model = LinearRegression()
# fitting of model
model.fit(x_train,y_train)
```

```python
# outcome of model's prediction
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# y_test: true mpg values
# y_pred: predicted mpg values from your model

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
# Calculate RMSE by taking the square root of MSE
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f} mpg")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f} mpg")
print(f"R² score: {r2:.2f}")
```

```python
accuarcy = model.score(x_test,y_test)
print('The accuarcy of model is  ',round(accuarcy,2)*100)
```
